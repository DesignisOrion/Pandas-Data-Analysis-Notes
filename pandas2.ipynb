```python
import pandas as pd
df = pd.read_csv('stocks.csv')

print(df)
```

# Filtering Data

```python
df.loc[df['SERIES'] == 'EQ'] # filter only the series showing EQ
```

```python
# multiple location conditions such as Series and Timestamp
df.loc[(df['SERIES'] == 'EQ') & (df['TIMESTAMP'] == '2017-06-28')]
# df.loc[(df['SERIES'] == 'EQ') | (df['TIMESTAMP'] == '2017-06-28')] # seperating the conditions using the "or" condition
# df.loc[(df['SERIES'] == 'EQ') & (df['TIMESTAMP'] == '2017-06-28') & (df['LOW'] > 40)] # seperating the conditions using the "or" also "and"
```

```python
# creating a new dataframe
new_df= df.loc[(df['SERIES'] == 'EQ') & (df['TIMESTAMP'] == '2017-06-28')]
```

```python
# Saving new dataFrame
new_df.to_csv('filtered.csv')
new_df
```

# Reset Indexes

```python
new_df = new_df.reset_index(drop=True, inplace=True) #drop will allow the old indexes to be removed from dataFrame and inplace keeps the dataframe the same
new_df
```
